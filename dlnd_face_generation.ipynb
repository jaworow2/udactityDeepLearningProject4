# Define the Model

A GAN is comprised of two adversarial networks, a discriminator and a generator.

## Discriminator

Your first task will be to define the discriminator. This is a convolutional classifier like you've built before, only without any maxpooling layers. To deal with this complex data, it's suggested you use a deep network with **normalization**. You are also allowed to create any helper functions that may be useful.

#### Exercise: Complete the Discriminator class
* The inputs to the discriminator are 32x32x3 tensor images
* The output should be a single value that will indicate whether a given image is real or fake


```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
# helper conv function
def conv(in_channels, out_channels, kernel_size, stride=2, padding=1, batch_norm=True):
    """Creates a convolutional layer, with optional batch normalization.
    
    Code Attribution:
        This is Udacity code.  No changes made to this function.
    """
    layers = []
    conv_layer = nn.Conv2d(in_channels, out_channels, 
                           kernel_size, stride, padding, bias=False)
    
    # append conv layer
    layers.append(conv_layer)

    if batch_norm:
        # append batchnorm layer
        layers.append(nn.BatchNorm2d(out_channels))
     
    # using Sequential container
    return nn.Sequential(*layers)
```

```python
class Discriminator(nn.Module):

    def __init__(self, conv_dim):
        """
        Initialize the Discriminator Module
        :param conv_dim: The depth of the first convolutional layer
        
        Code Attribution:
            This function contains code that was updated and leveraged from the 
            Udacity Deep Learning Nano Degree Training material.        
        """
        super(Discriminator, self).__init__()
        
        # complete init function
        
        # acquire depth of first layer
        self.conv_dim = conv_dim
        
        # define convolutionaly layers
        
        # (32x32x3) -> (16x16x64)
        self.conv1 = conv(in_channels=3, out_channels=conv_dim, kernel_size=4, stride=2, batch_norm=False)
        # (16x16x64) -> (8x8x128)
        self.conv2 = conv(in_channels=conv_dim, out_channels=conv_dim*2, kernel_size=4, stride=2)
        # (8x8x128) -> (4x4x256)
        self.conv3 = conv(in_channels=conv_dim*2, out_channels=conv_dim*4, kernel_size=4, stride=2)  
        
        # define fully connected layer
        self.fc1 = nn.Linear(conv_dim*4*4*4, 1)

    def forward(self, x):
        """
        Forward propagation of the neural network
        :param x: The input to the neural network     
        :return: Discriminator logits; the output of the neural network
        
        Code Attribution:
            This function contains code that was updated and leveraged from the 
            Udacity Deep Learning Nano Degree Training material.              
        """
        # define feedforward behavior
        
        # pass image through convolution layers and leaky relu activation function
        x = self.conv1(x)
        x = F.leaky_relu(x)
        x = self.conv2(x)
        x = F.leaky_relu(x)
        x = self.conv3(x)
        x = F.leaky_relu(x)
        
        # convert to flatten feature vector for fully connected layer
        x = x.view(-1, self.conv_dim*4*4*4)
        
        # pass through fully connected layer
        x = self.fc1(x)
        
        return x


"""
DON'T MODIFY ANYTHING IN THIS CELL THAT IS BELOW THIS LINE
"""
tests.test_discriminator(Discriminator)
```

## Generator

The generator should upsample an input and generate a *new* image of the same size as our training data `32x32x3`. This should be mostly transpose convolutional layers with normalization applied to the outputs.

#### Exercise: Complete the Generator class
* The inputs to the generator are vectors of some length `z_size`
* The output should be a image of shape `32x32x3`

```python
# helper deconv function
def deconv(in_channels, out_channels, kernel_size, stride=2, padding=1, batch_norm=True):
    """Creates a transposed-convolutional layer, with optional batch normalization.
    
    Code Attribution:
        This is Udacity code.  No changes made to this function.
    """
    ## TODO: Complete this function
    ## create a sequence of transpose + optional batch norm layers

    layers = []
    deconv_layer = nn.ConvTranspose2d(in_channels, out_channels, kernel_size, 
                                      stride, padding, bias=False)
    
    # append deconv layers
    layers.append(deconv_layer)
    
    if batch_norm:
        # append batchnorm layer
        layers.append(nn.BatchNorm2d(out_channels))
    
    # sequential container
    return nn.Sequential(*layers)
```

```python
class Generator(nn.Module):
    
    def __init__(self, z_size, conv_dim):
        """
        Initialize the Generator Module
        :param z_size: The length of the input latent vector, z
        :param conv_dim: The depth of the inputs to the *last* transpose convolutional layer
        
        Code Attribution:
            This function contains code that was updated and leveraged from the 
            Udacity Deep Learning Nano Degree Training material.        
        """
        super(Generator, self).__init__()

        # complete init function
        
        # acquire depth of first layer
        self.conv_dim = conv_dim
        
        # define fully connected layers
        self.fc1 = nn.Linear(z_size, conv_dim*8*2*2)
        
        # transpose convoluational layers
        # (2x2x512) -> (4x4x256)
        self.deconv1 = deconv(in_channels=conv_dim*8, out_channels=conv_dim*4, kernel_size=4, stride=2)
        # (4x4x256) -> (8x8x128)
        self.deconv2 = deconv(in_channels=conv_dim*4, out_channels=conv_dim*2, kernel_size=4, stride=2)
        # (8x8x128) -> (16x16x64) 
        self.deconv3 = deconv(in_channels=conv_dim*2, out_channels=conv_dim, kernel_size=4, stride=2)
        # (16x16x64) -> (32x32x3)
        self.deconv4 = deconv(in_channels=conv_dim, out_channels=3, kernel_size=4, stride=2, batch_norm=False)

    def forward(self, x):
        """
        Forward propagation of the neural network
        :param x: The input to the neural network     
        :return: A 32x32x3 Tensor image as output
        
        Code Attribution:
            This function contains code that was updated and leveraged from the 
            Udacity Deep Learning Nano Degree Training material.              
        """
        # define feedforward behavior
        
        # pass through fully connected layer
        x = self.fc1(x)
        
        # reshape feature vector to matrix for deconvolutional layers
        x = x.view(-1, self.conv_dim*8, 2, 2)
        
        # pass through deconvolutional layers, relu activation, and the final tanh expected output function
        x = self.deconv1(x)
        x = F.relu(x)
        x = self.deconv2(x)
        x = F.relu(x)
        x = self.deconv3(x)
        x = F.relu(x)
        x = self.deconv4(x)        
        x = torch.tanh(x)
        
        return x

"""
DON'T MODIFY ANYTHING IN THIS CELL THAT IS BELOW THIS LINE
"""
tests.test_generator(Generator)
```

```python
from torch.nn import init
def weights_init_normal(m):
    """
    Applies initial weights to certain layers in a model .
    The weights are taken from a normal distribution 
    with mean = 0, std dev = 0.02.
    :param m: A module or layer in a network    
    
    Code Attribution:
        This function contains code that was updated and leveraged from the 
        https://github.com/junyanz/pytorch-CycleGAN-and-pix2pix/blob/master/models/networks.py
    """
    # classname will be something like:
    # `Conv`, `BatchNorm2d`, `Linear`, etc.
    classname = m.__class__.__name__
    
    # TODO: Apply initial weights to convolutional and linear layers
    
    # convolutional or linear layers
    if (classname.find('Conv') != -1 or classname.find('Linear') != -1):
        init.normal_(m.weight.data, mean=0.0, std=.02)
    # BatchNorm2d layer
    elif (classname.find('BatchNorm2d') != -1):
        init.normal_(m.weight.data, mean=0.0, std=.02)
        init.constant_(m.bias.data, 0.0)
    # other
    else:
        None
    
```

## Discriminator and Generator Losses

Now we need to calculate the losses for both types of adversarial networks.

### Discriminator Losses

> * For the discriminator, the total loss is the sum of the losses for real and fake images, `d_loss = d_real_loss + d_fake_loss`. 
* Remember that we want the discriminator to output 1 for real images and 0 for fake images, so we need to set up the losses to reflect that.


### Generator Loss

The generator loss will look similar only with flipped labels. The generator's goal is to get the discriminator to *think* its generated images are *real*.

#### Exercise: Complete real and fake loss functions

**You may choose to use either cross entropy or a least squares error loss to complete the following `real_loss` and `fake_loss` functions.**

```python
def real_loss(D_out):
    '''Calculates how close discriminator outputs are to being real.
       param, D_out: discriminator logits
       return: real loss
    
    Code Attribution:
        This function contains code that was updated and leveraged from the 
        Udacity Deep Learning Nano Degree Training material.
    '''    
    # acquire batch_size
    batch_size = D_out.size(0)
    
    # real_loss expects a label value of 1
    labels = torch.ones(batch_size)
    if train_on_gpu:
        labels = labels.cuda()
    
    # define the binary cross entropy loss function
    criterion = nn.BCEWithLogitsLoss()
    # calculates loss
    loss = criterion(D_out.squeeze(), labels)
    return loss

def fake_loss(D_out):
    '''Calculates how close discriminator outputs are to being fake.
       param, D_out: discriminator logits
       return: fake loss
    
    Code Attribution:
        This function contains code that was updated and leveraged from the 
        Udacity Deep Learning Nano Degree Training material.
    '''
    # acquire batch_size
    batch_size = D_out.size(0)
    
    # fake_loss expects a label value of 0
    labels = torch.zeros(batch_size)
    if train_on_gpu:
        labels = labels.cuda()    
    
    # define the binary cross entropy loss function
    criterion = nn.BCEWithLogitsLoss()
    # calculates loss
    loss = criterion(D_out.squeeze(), labels)
    return loss
```

## Training

Training will involve alternating between training the discriminator and the generator. You'll use your functions `real_loss` and `fake_loss` to help you calculate the discriminator losses.

* You should train the discriminator by alternating on real and fake images
* Then the generator, which tries to trick the discriminator and should have an opposing loss function


#### Saving Samples

You've been given some code to print out some loss statistics and save some generated "fake" samples.

#### Exercise: Complete the training function

Keep in mind that, if you've moved your models to GPU, you'll also have to move any model inputs to GPU.

```python
def train(D, G, n_epochs, print_every=50):
    '''Trains adversarial networks for some number of epochs
       param, D: the discriminator network
       param, G: the generator network
       param, n_epochs: number of epochs to train for
       param, print_every: when to print and record the models' losses
       return: D and G losses
       
       Code Attribution:
           This function contains code that was updated and leveraged from the 
           Udacity Deep Learning Nano Degree Training material.
       '''
    
    # move models to GPU
    if train_on_gpu:
        D.cuda()
        G.cuda()

    # keep track of loss and generated, "fake" samples
    samples = []
    losses = []

    # Get some fixed data for sampling. These are images that are held
    # constant throughout training, and allow us to inspect the model's performance
    sample_size=16
    fixed_z = np.random.uniform(-1, 1, size=(sample_size, z_size))
    fixed_z = torch.from_numpy(fixed_z).float()
    # move z to GPU if available
    if train_on_gpu:
        fixed_z = fixed_z.cuda()

    # epoch training loop
    for epoch in range(n_epochs):

        # batch training loop
        for batch_i, (real_images, _) in enumerate(celeba_train_loader):

            batch_size = real_images.size(0)
            real_images = scale(real_images)

            # ===============================================
            #         YOUR CODE HERE: TRAIN THE NETWORKS
            # ===============================================
            
            # 1. Train the discriminator on real and fake images
            # zero out discriminator gradients
            d_optimizer.zero_grad()
            
            # move images to gpu if available
            if train_on_gpu:
                real_images = real_images.cuda()
            
            # compute discriminator losses on real images
            d_real_out = D(real_images)
            d_real_loss = real_loss(d_real_out)
            
            # compute discriminator losses on fake images
            
            # generate random z vector for fake images
            z = np.random.uniform(-1, 1, size=(batch_size, z_size))
            z = torch.from_numpy(z).float()
            
            # move fake images to gpu if availabel
            if train_on_gpu:
                z = z.cuda()
            
            # have generator create fake images from random z vector
            fake_images = G(z)
            
            # compute discriminator losses on fake images
            d_fake_out = D(fake_images)
            d_fake_loss = fake_loss(d_fake_out)
            
            # add up losses, perform backpropagation, and update weights
            d_loss = d_real_loss + d_fake_loss
            d_loss.backward()
            d_optimizer.step()

            # 2. Train the generator with an adversarial loss
            
            # zero out generator gradients
            g_optimizer.zero_grad()
            
            # generate random z vector for fake images
            z = np.random.uniform(-1, 1, size=(batch_size, z_size))
            z = torch.from_numpy(z).float()
            
            # move fake images to gpu if availabel
            if train_on_gpu:
                z = z.cuda()            

            # have generator create fake images from random z vector
            fake_images = G(z)
            
            # compute discriminator losses on generator fake images, 
            # but this time apply real loss function
            d_fake_out = D(fake_images)
            g_loss = real_loss(d_fake_out)
            
            # perform backpropagation and update weights
            g_loss.backward()
            g_optimizer.step()
            
            
            # ===============================================
            #              END OF YOUR CODE
            # ===============================================

            # Print some loss stats
            if batch_i % print_every == 0:
                # append discriminator loss and generator loss
                losses.append((d_loss.item(), g_loss.item()))
                # print discriminator and generator loss
                print('Epoch [{:5d}/{:5d}] | d_loss: {:6.4f} | g_loss: {:6.4f}'.format(
                        epoch+1, n_epochs, d_loss.item(), g_loss.item()))


        ## AFTER EACH EPOCH##    
        # this code assumes your generator is named G, feel free to change the name
        # generate and save sample, fake images
        G.eval() # for generating samples
        with torch.no_grad():
            samples_z = G(fixed_z)
            samples_z = samples_z.detach().cpu()
            samples.append(samples_z)
        G.train() # back to training mode

    # Save training generator samples
    with open('train_samples.pkl', 'wb') as f:
        pkl.dump(samples, f)
    
    # finally return losses
    return losses
```

To improve the generated images further, I would pursue creating a deeper model combined with a dynamic learning rate that decays over time.
